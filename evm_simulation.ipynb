# Project for simulating an EVM 
This is a sample simulating EVM or electronic voting machine by code.
It takes votes of n people and then shows the result at the end.

```python
import time

n = int(input("Enter the number of people voting: "))
parties = {
    0:"National Reform Party",
    1:"Alliance Party",
    2:"Socialist League",
    3:"Communist Alliance",
    4:"Secular Democratic Party",
    5:"Socialist Labour Party",
    6:"National Liberal Alliance",
    7:"Left Radical Party",
    8:"United Left"
    
}

list1 = [0]*9

for j,k in parties.items():
        print(j, ": ", k)

for i in range(n):
    k = int(input("\nPlease press the number on the machine with accordance to the party you wish to vote for: "))
    print("\nThank you for voting!")
    list1[k] += 1
    
time.sleep(1.7)

k=0
print("\nThe results are: ")
for i in parties.values():
    print(i, f" got {list1[k]} votes")
    k+=1
```
